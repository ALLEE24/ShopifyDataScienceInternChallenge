```python
#ANSWER TO QUESTION 1a:
#
#Looking at the data, there are 17 outliers in a set of 5000 (0.34$) skewing the data. 
#Each one is by the same person (ID 67) from the same shop (ID 42) for 704 000$ and 2000 pairs of sneakers.
#The next highest amount of orders is 6.
#This is highly irregular activity relative to the rest of the data set, and may indicate an accounting error or even possibly credit card fraud and/or money laundering.
#In any case, these orders, specifically their high volume, are skewing the dataset.
#There are other high value orders >$100 000, but these are all from different users for <6 pairs of sneakers.
#These are less irregular, and may indicate high value collector/limited edition sneakers.
#For the purposes of finding a more accurate AOV, I will remove the irregular values (2000 items)
```

```python
no_outliers = data.copy()
no_outliers = no_outliers[no_outliers.total_items < 10]
no_outliers = no_outliers.sort_values('total_items', ascending = False)
no_outliers.head(30)
```
